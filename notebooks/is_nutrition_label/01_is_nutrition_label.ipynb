## Gathering Data

Our Bing Images API allows you to scrape SERP results from Bing Images. The API is accessed through the following endpoint: /search?engine=bing_images.

A user may query the following: https://serpapi.com/search?engine=bing_images utilizing a GET request. Head to the playground for a live and interactive demo.

```python
from serpapi import GoogleSearch
import requests
from pathlib import Path
from time import sleep

def download_nutrition_labels(query, num_images=200):
    from pathlib import Path

    # Base path
    save_dir = Path("../../data/is_nutrition_label/")

    # Add subfolder
    path = query.replace(' ', '_')
    full_path = save_dir / path  # This creates: ../../data/is_nutrition_label/nutrition_facts_label

    # Create the directory
    full_path.mkdir(parents=True, exist_ok=True)
    
    params = {
        "engine": "google_images",  # or "bing_images"
        "q": query,
        "api_key": SERPAPI_API_KEY,
        "num": 100
    }
    
    search = GoogleSearch(params)
    results = search.get_dict()
    
    downloaded = 0
    for idx, img in enumerate(results.get("images_results", [])):
        if downloaded >= num_images:
            break
            
        try:
            img_url = img.get('original') or img.get('thumbnail')
            response = requests.get(img_url, timeout=10, headers={'User-Agent': 'Mozilla/5.0'})
            
            if response.status_code == 200:
                file_path = full_path / f"nutrition_{downloaded:04d}.jpg"
                with open(file_path, 'wb') as f:
                    f.write(response.content)
                print(f"✅ Downloaded {downloaded+1}/{num_images}")
                downloaded += 1
                sleep(0.5)  # Be respectful
        except Exception as e:
            print(f"❌ Failed {idx}: {e}")
    
    return downloaded

# Download different types
download_nutrition_labels("nutrition facts label")
download_nutrition_labels("processed food")
```
